# Experiment-level parameter recovery audit

Question: does the current browser adaptive workflow recover known simulated participants, and how do Stan MI ADO, Stan random designs, and Quest+ compare on the same experiment grid?

This notebook drives the actual jsPsych experiment page in `simulate=data-only` mode. It is an explainer/audit template, not CI and not a definitive efficiency study. The table used for analysis is normalized to one row per experiment, model, strategy label, simulation profile, seed, trial, and parameter so the same notebook shape can cover future models with different parameter counts.

The comparison arms answer slightly different questions. Stan MI ADO is the main workflow: Stan/WASM posterior updates with mutual-information design selection. Stan random keeps the same Stan/WASM posterior updates but samples designs randomly from the same grid, isolating the value of adaptive design selection. Quest+ is a separate discrete-grid adaptive controller and posterior, so it is an alternative-controller benchmark rather than a pure design-strategy ablation.

The rendered reports use fixed randomized settings files so the empirical error bands and true-vs-recovered plots are reproducible. By default, the parameter cell still runs the compact delay-discounting profile audit. To run a rendered report, execute the notebook with papermill and set `SETTINGS_PATH` to a JSON settings file such as `examples/parameter_recovery/delay_discounting_random_recovery_settings.json` or `examples/parameter_recovery/line_length_discrimination_random_recovery_settings.json`.


```python
# Parameters
SETTINGS_PATH = None
EXPERIMENT_PATH = "experiments/delay_discounting/index.html"
MODEL_ID = "hyperbolic"
CHOICE_TASK_FIELD = "task"
CHOICE_TASK = "delay_discounting"
TRIAL_NUMBER_FIELD = "trial_number"
DESIGN_FIELDS = ["r_ss", "t_ss", "r_ll", "t_ll"]
RUN_SPECS = [
    {"strategy": "ado", "query": {"controller": "stan", "strategy": "ado"}},
    {"strategy": "random", "query": {"controller": "stan", "strategy": "random"}},
    {"strategy": "quest_plus", "query": {"controller": "quest_plus"}},
]
STRATEGIES = [spec["strategy"] for spec in RUN_SPECS]
SIMULATION_PROFILES = [
    {"profile_id": "k_1e-4", "sweep": "k", "params": {"k": 1e-4, "tau": 2.5}},
    {"profile_id": "k_1e-3", "sweep": "k", "params": {"k": 1e-3, "tau": 2.5}},
    {"profile_id": "k_1e-2", "sweep": "k", "params": {"k": 1e-2, "tau": 2.5}},
    {"profile_id": "tau_0.5", "sweep": "tau", "params": {"k": 5e-3, "tau": 0.5}},
    {"profile_id": "tau_2.5", "sweep": "tau", "params": {"k": 5e-3, "tau": 2.5}},
    {"profile_id": "tau_5.0", "sweep": "tau", "params": {"k": 5e-3, "tau": 5.0}},
]
SEEDS = [101, 102, 103, 104, 105]
PARAMETERS = [
    {"name": "k", "scale": "log10", "label": "Discount rate k"},
    {"name": "tau", "scale": "linear", "label": "Choice sensitivity tau"},
]
SELECTED_PARAMETERS = ["k", "tau"]
RUN_QUICK = False
DIRECTIONAL_CHECKS = []
INCLUDE_RAW = False
```

```python
# Parameters
SETTINGS_PATH = "examples/parameter_recovery/delay_discounting_random_recovery_settings.json"
```

## Run the browser audit

The helper starts a local static server, injects `window.__JSPSYCH_ADO_RUN_CONFIG__`, opens the experiment with `simulate=data-only`, and reads the displayed jsPsych JSON. The notebook calls the helper in a subprocess so Playwright runs outside the notebook kernel event loop.

```python
import json
import math
import subprocess
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

REPO_ROOT = Path.cwd().resolve()
while not (REPO_ROOT / EXPERIMENT_PATH).exists() and REPO_ROOT != REPO_ROOT.parent:
    REPO_ROOT = REPO_ROOT.parent

HELPER_PATH = REPO_ROOT / "examples/parameter_recovery/parameter_recovery_browser.py"
if not HELPER_PATH.exists():
    raise FileNotFoundError(HELPER_PATH)

if SETTINGS_PATH:
    settings_path = Path(SETTINGS_PATH)
    if not settings_path.is_absolute():
        settings_path = REPO_ROOT / settings_path
    external_settings = json.loads(settings_path.read_text())
    EXPERIMENT_PATH = external_settings.get("experiment_path", EXPERIMENT_PATH)
    MODEL_ID = external_settings.get("model_id", MODEL_ID)
    CHOICE_TASK_FIELD = external_settings.get("choice_task_field", CHOICE_TASK_FIELD)
    CHOICE_TASK = external_settings.get("choice_task", CHOICE_TASK)
    TRIAL_NUMBER_FIELD = external_settings.get("trial_number_field", TRIAL_NUMBER_FIELD)
    DESIGN_FIELDS = external_settings.get("design_fields", DESIGN_FIELDS)
    RUN_SPECS = external_settings.get("run_specs", RUN_SPECS)
    STRATEGIES = [spec["strategy"] for spec in RUN_SPECS]
    SIMULATION_PROFILES = external_settings.get("simulation_profiles", SIMULATION_PROFILES)
    SEEDS = external_settings.get("seeds", SEEDS)
    PARAMETERS = external_settings.get("parameters", PARAMETERS)
    SELECTED_PARAMETERS = external_settings.get("selected_parameters", [param["name"] for param in PARAMETERS])
    DIRECTIONAL_CHECKS = external_settings.get("directional_checks", DIRECTIONAL_CHECKS)

profiles = SIMULATION_PROFILES[:1] if RUN_QUICK else SIMULATION_PROFILES
seeds = SEEDS[:1] if RUN_QUICK else SEEDS
settings = {
    "experiment_path": EXPERIMENT_PATH,
    "model_id": MODEL_ID,
    "run_specs": RUN_SPECS,
    "simulation_profiles": profiles,
    "seeds": seeds,
    "parameters": PARAMETERS,
    "choice_task_field": CHOICE_TASK_FIELD,
    "choice_task": CHOICE_TASK,
    "trial_number_field": TRIAL_NUMBER_FIELD,
    "design_fields": DESIGN_FIELDS,
    "directional_checks": DIRECTIONAL_CHECKS,
    "repo_root": str(REPO_ROOT),
}
cmd = [sys.executable, str(HELPER_PATH), "--settings-json", json.dumps(settings)]
if INCLUDE_RAW:
    cmd.append("--include-raw")
completed = subprocess.run(cmd, check=True, capture_output=True, text=True)
payload = json.loads(completed.stdout)
metadata = payload["metadata"]
parameter_rows = pd.DataFrame(payload["rows"])
display(pd.DataFrame([{
    "experiment_path": metadata["experiment_path"],
    "model_id": metadata["model_id"],
    "strategies": ", ".join(metadata["strategies"]),
    "profiles": len(metadata["simulation_profiles"]),
    "seeds": len(metadata["seeds"]),
    "trial_count": metadata["trial_count"],
    "settings_path": SETTINGS_PATH or "inline",
}]))
parameter_rows.head()
```

```python
required_columns = {
    "experiment_path",
    "model_id",
    "strategy",
    "profile_id",
    "seed",
    "trial_number",
    "parameter",
    "true_value",
    "post_mean",
    "post_sd",
    "error",
    "abs_error",
}
missing_columns = required_columns.difference(parameter_rows.columns)
if missing_columns:
    raise ValueError(f"Missing normalized columns: {sorted(missing_columns)}")

observed_strategies = set(parameter_rows["strategy"].unique())
expected_strategies = set(STRATEGIES)
if observed_strategies != expected_strategies:
    raise ValueError(f"Expected strategies {sorted(expected_strategies)}, got {sorted(observed_strategies)}")

trial_counts = (
    parameter_rows
    .groupby(["strategy", "profile_id", "seed"], as_index=False)["trial_number"]
    .nunique()
    .rename(columns={"trial_number": "n_trials"})
)
display(trial_counts.head())
```

## Final-trial recovery

Each row below summarizes the last available trial for one parameter and strategy. Error is computed on the parameter's configured scale: log10 for parameters such as `k`, raw units for linear parameters. Rank correlation is computed between true values and final posterior means on the same transformed plotting scale.


```python
def transform_value(value, scale):
    if pd.isna(value):
        return math.nan
    value = float(value)
    if scale == "log10":
        return math.log10(value) if value > 0 else math.nan
    return value


def rank_correlation(group, scale):
    transformed = pd.DataFrame({
        "true_plot": [transform_value(value, scale) for value in group["true_value"]],
        "post_plot": [transform_value(value, scale) for value in group["post_mean"]],
    }).dropna()
    if len(transformed) < 2:
        return math.nan
    return transformed["true_plot"].rank().corr(transformed["post_plot"].rank())


final_rows = (
    parameter_rows
    .sort_values("trial_number")
    .groupby(["strategy", "profile_id", "seed", "parameter"], as_index=False)
    .tail(1)
)

summary_rows = []
for (parameter, label, scale, strategy), group in final_rows.groupby(
    ["parameter", "parameter_label", "parameter_scale", "strategy"],
    sort=False,
):
    abs_error = group["abs_error"].astype(float)
    post_sd = group["post_sd"].astype(float)
    summary_rows.append({
        "parameter": parameter,
        "parameter_label": label,
        "parameter_scale": scale,
        "strategy": strategy,
        "median_abs_error": abs_error.median(),
        "q25_abs_error": abs_error.quantile(0.25),
        "q75_abs_error": abs_error.quantile(0.75),
        "iqr_abs_error": abs_error.quantile(0.75) - abs_error.quantile(0.25),
        "median_post_sd": post_sd.median(),
        "rank_correlation": rank_correlation(group, scale),
        "n_runs": len(group),
    })

final_summary = (
    pd.DataFrame(summary_rows)
    .sort_values(["parameter", "strategy"])
    .reset_index(drop=True)
)
display(final_summary)
```

## Profile-level posterior means

The table below keeps representative simulation profiles visible. For randomized reports this table can be long, so the rendered notebook displays the first rows while the full normalized table remains available in `parameter_rows`.


```python
profile_summary = (
    final_rows
    .groupby(["parameter", "parameter_label", "parameter_scale", "strategy", "profile_id"], as_index=False)
    .agg(
        true_value=("true_value", "median"),
        median_post_mean=("post_mean", "median"),
        median_abs_error=("abs_error", "median"),
        median_post_sd=("post_sd", "median"),
        n_runs=("post_mean", "size"),
    )
    .sort_values(["parameter", "strategy", "true_value", "profile_id"])
)
display(profile_summary.head(24))
```

## Directional checks

Some recovery audits need a model-specific sanity check: when one simulation profile has a higher true value for a parameter than another, does the final posterior mean recover the same ordering? These checks are optional and come from the settings JSON.


```python
directional_rows = []
for check in DIRECTIONAL_CHECKS:
    parameter_name = check["parameter"]
    lower_profile = check["lower_profile"]
    higher_profile = check["higher_profile"]
    for strategy, strategy_rows in final_rows[final_rows["parameter"] == parameter_name].groupby("strategy"):
        lower_values = strategy_rows[strategy_rows["profile_id"] == lower_profile]["post_mean"]
        higher_values = strategy_rows[strategy_rows["profile_id"] == higher_profile]["post_mean"]
        if lower_values.empty or higher_values.empty:
            continue
        lower_post_mean = lower_values.median()
        higher_post_mean = higher_values.median()
        directional_rows.append({
            "parameter": parameter_name,
            "strategy": strategy,
            "lower_profile": lower_profile,
            "higher_profile": higher_profile,
            "lower_post_mean": lower_post_mean,
            "higher_post_mean": higher_post_mean,
            "difference": higher_post_mean - lower_post_mean,
            "passed": higher_post_mean > lower_post_mean,
            "description": check.get("description", ""),
        })

directional_results = pd.DataFrame(directional_rows)
if directional_results.empty:
    print("No directional checks configured.")
else:
    display(directional_results)
```

## Error and uncertainty trajectories

Trajectory summaries keep parameters long rather than putting each parameter in its own column. Lines show the median across simulated runs. Ribbons show empirical variation across true profile and response seed: the darker band is the interquartile range, and the lighter outer band is the 10th-90th percentile range where shown.


```python
trajectory = (
    parameter_rows
    .groupby(["parameter", "parameter_label", "parameter_scale", "strategy", "trial_number"], as_index=False)
    .agg(
        median_abs_error=("abs_error", "median"),
        q10_abs_error=("abs_error", lambda values: values.quantile(0.10)),
        q25_abs_error=("abs_error", lambda values: values.quantile(0.25)),
        q75_abs_error=("abs_error", lambda values: values.quantile(0.75)),
        q90_abs_error=("abs_error", lambda values: values.quantile(0.90)),
        mean_abs_error=("abs_error", "mean"),
        median_post_sd=("post_sd", "median"),
        q25_post_sd=("post_sd", lambda values: values.quantile(0.25)),
        q75_post_sd=("post_sd", lambda values: values.quantile(0.75)),
        mean_post_sd=("post_sd", "mean"),
    )
)
display(trajectory.head())
```

```python
def plot_band_by_parameter(
    data,
    center,
    lower,
    upper,
    ylabel,
    selected_parameters,
    outer_lower=None,
    outer_upper=None,
):
    for parameter_name in selected_parameters:
        plot_data = data[data["parameter"] == parameter_name]
        if plot_data.empty:
            continue
        label = plot_data["parameter_label"].iloc[0]
        fig, ax = plt.subplots(figsize=(7, 4))
        for strategy, strategy_data in plot_data.groupby("strategy"):
            strategy_data = strategy_data.sort_values("trial_number")
            x = strategy_data["trial_number"].to_numpy(dtype=float)
            y = strategy_data[center].to_numpy(dtype=float)
            line = ax.plot(x, y, label=strategy, linewidth=1.8)[0]
            color = line.get_color()
            if outer_lower and outer_upper:
                ax.fill_between(
                    x,
                    strategy_data[outer_lower].to_numpy(dtype=float),
                    strategy_data[outer_upper].to_numpy(dtype=float),
                    color=color,
                    alpha=0.10,
                    linewidth=0,
                )
            ax.fill_between(
                x,
                strategy_data[lower].to_numpy(dtype=float),
                strategy_data[upper].to_numpy(dtype=float),
                color=color,
                alpha=0.22,
                linewidth=0,
            )
        ax.set_title(label)
        ax.set_xlabel("Trial")
        ax.set_ylabel(ylabel)
        ax.legend(title="Strategy")
        ax.grid(alpha=0.25)
        plt.show()


plot_band_by_parameter(
    trajectory,
    "median_abs_error",
    "q25_abs_error",
    "q75_abs_error",
    "Absolute error",
    SELECTED_PARAMETERS,
    outer_lower="q10_abs_error",
    outer_upper="q90_abs_error",
)
plot_band_by_parameter(
    trajectory,
    "median_post_sd",
    "q25_post_sd",
    "q75_post_sd",
    "Posterior SD",
    SELECTED_PARAMETERS,
)
```

## Final-trial distributions and true-vs-recovered checks

Final-trial plots show how recovery varies across simulated true values and response seeds. The true-vs-recovered scatter uses each parameter's configured plotting scale and includes an identity line; points close to the line indicate better recovery.


```python
def plot_final_error_distribution(data, selected_parameters):
    for parameter_name in selected_parameters:
        plot_data = data[data["parameter"] == parameter_name]
        if plot_data.empty:
            continue
        label = plot_data["parameter_label"].iloc[0]
        strategies = [strategy for strategy in STRATEGIES if strategy in set(plot_data["strategy"])]
        values_by_strategy = [
            plot_data[plot_data["strategy"] == strategy]["abs_error"].astype(float).tolist()
            for strategy in strategies
        ]
        fig, ax = plt.subplots(figsize=(7, 4))
        positions = list(range(1, len(strategies) + 1))
        ax.boxplot(values_by_strategy, positions=positions, showfliers=False)
        ax.set_xticks(positions)
        ax.set_xticklabels(strategies)
        for position, values in zip(positions, values_by_strategy):
            offsets = [((index % 9) - 4) * 0.018 for index in range(len(values))]
            ax.scatter(
                [position + offset for offset in offsets],
                values,
                color="black",
                alpha=0.18,
                s=9,
                linewidths=0,
            )
        ax.set_title(label)
        ax.set_xlabel("Strategy")
        ax.set_ylabel("Final absolute error")
        ax.grid(axis="y", alpha=0.25)
        plt.show()


def plot_true_vs_recovered(data, selected_parameters):
    for parameter_name in selected_parameters:
        plot_data = data[data["parameter"] == parameter_name].copy()
        if plot_data.empty:
            continue
        label = plot_data["parameter_label"].iloc[0]
        scale = plot_data["parameter_scale"].iloc[0]
        plot_data["true_plot"] = [transform_value(value, scale) for value in plot_data["true_value"]]
        plot_data["post_plot"] = [transform_value(value, scale) for value in plot_data["post_mean"]]
        plot_data = plot_data.dropna(subset=["true_plot", "post_plot"])
        if plot_data.empty:
            continue
        fig, ax = plt.subplots(figsize=(5.5, 5))
        for strategy, strategy_data in plot_data.groupby("strategy"):
            ax.scatter(
                strategy_data["true_plot"],
                strategy_data["post_plot"],
                alpha=0.45,
                s=18,
                label=strategy,
            )
        axis_min = min(plot_data["true_plot"].min(), plot_data["post_plot"].min())
        axis_max = max(plot_data["true_plot"].max(), plot_data["post_plot"].max())
        padding = (axis_max - axis_min) * 0.05 if axis_max > axis_min else 1
        axis_min -= padding
        axis_max += padding
        ax.plot([axis_min, axis_max], [axis_min, axis_max], color="black", linestyle="--", linewidth=1)
        ax.set_xlim(axis_min, axis_max)
        ax.set_ylim(axis_min, axis_max)
        axis_label = f"log10({parameter_name})" if scale == "log10" else parameter_name
        ax.set_title(label)
        ax.set_xlabel(f"True {axis_label}")
        ax.set_ylabel(f"Recovered {axis_label}")
        ax.legend(title="Strategy")
        ax.grid(alpha=0.25)
        plt.show()


plot_final_error_distribution(final_rows, SELECTED_PARAMETERS)
plot_true_vs_recovered(final_rows, SELECTED_PARAMETERS)

display(final_summary.pivot_table(
    index=["parameter", "parameter_label", "parameter_scale"],
    columns="strategy",
    values="median_abs_error",
))
```

## Interpretation notes

Read this as an audit of the current browser workflow: the task page, jsPsych simulation behavior, Stan/WASM posterior updates, Quest+ posterior updates, and design-selection policy are all in the loop. Error bands are empirical variation across simulated participants, not posterior credible intervals. Posterior uncertainty is shown separately through posterior-SD bands. Stronger adaptive performance should show up as lower final error, faster median error reduction, and true-vs-recovered points closer to the identity line relative to the random baseline. Quest+ uses a discrete parameter grid, while the Stan controller summarizes continuous posterior samples, so compare the trajectories as workflow evidence rather than as a definitive efficiency result. Remaining caveats are model-specific, grid-specific, and simulation-size-specific; this notebook should be rerun with broader or task-specific profile settings before making efficiency claims.
